```python
import pandas as pd
```

```python
data = pd.read_excel('Mentees.xlsx')
data.head()
```

```python
data = data[['In what city do you live?', 'Career field interest', 'What industries interest you?', 'What are you looking to gain from mentorship?', 'What skills, interests, hobbies, or talents would you like to share with your mentor?', 'What are the top 1-3 priorities you would like us to consider when matching you to a mentor?', 'What would you want your mentor to know about you before you meet?']]
data.head()
```

```python
cities = data['In what city do you live?'].str.lower()
cities = [city.split(',')[0].replace(' ', '') for city in cities]
```

```python
data['city'] = cities
data = data.drop(['In what city do you live?'], axis=1)
data.head()
```

```python
data = data.rename(columns={'Career field interest': 'career_interest', 'What industries interest you?': 'industries', 'What are you looking to gain from mentorship?': 'gain', 'What skills, interests, hobbies, or talents would you like to share with your mentor?': 'skills_hobbies', 'What would you want your mentor to know about you before you meet?': 'about_you', 'What are the top 1-3 priorities you would like us to consider when matching you to a mentor?': 'priorities'})
```

```python
import numpy as np

data = data.replace(np.nan, '')
```

```python
data
```

```python
from rake_nltk import Rake

for col in data.drop(['city'], axis=1):
    k = []
    for row in data[col]:
        r = Rake()
        r.extract_keywords_from_text(row)

        key_words_dict_scores = r.get_word_degrees()

        k.append(list(key_words_dict_scores.keys()))
    data[col] = k
```

```python
def make_bag_of_words(df):
    df['bag_of_words'] = ''
    columns = df.columns
    for index, row in df.iterrows():
        words = ''
        for col in columns:
            if type(row[col]) != list:
                words = words + row[col] + ' '
            else:
                words = words + ' '.join(row[col]) + ' '
        row['bag_of_words'] = words

    df.drop(columns=[col for col in df.columns if col != 'bag_of_words' and col != 'Full Name'],
            inplace=True)
```

```python
make_bag_of_words(data)
data["bag_of_words"] = data["bag_of_words"].str.lower()
data
```
